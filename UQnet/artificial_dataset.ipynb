```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 200
# X = np.linspace(0.1, 0.9, N)
X = np.sort(np.linspace(0.1, 0.9, N) + np.random.uniform(-0.1, 0.1, size=N))
Y = np.sin(8*X) + 0.01*np.random.normal(loc=0, scale=14, size=N)
```

```python
plt.plot(X, Y, ".")
plt.show()
```

```python
np.save("datasets/X_arti.npy", X)
np.save("datasets/Y_arti.npy", Y)
```
